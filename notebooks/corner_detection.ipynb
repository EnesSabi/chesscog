```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

from chesscog.corner_detection.detect_corners import (
    resize_image,
    detect_edges,
    detect_lines,
    cluster_horizontal_and_vertical_lines,
    eliminate_similar_lines,
    get_intersection_points,
    _choose_from_range,
    _fix_negative_rho_in_hesse_normal_form,
    compute_homography,
    warp_points,
    discard_outliers,
    quantize_points,
    compute_transformation_matrix,
    compute_horizontal_borders,
    compute_vertical_borders
)
from chesscog.corner_detection.visualise import draw_lines
from chesscog.utils import sort_corner_points
from chesscog.utils.coordinates import from_homogenous_coordinates, to_homogenous_coordinates
from recap import URI, CfgNode as CN
```

```python
cfg = CN.load_yaml_with_base(URI("config://corner_detection.yaml"))
#cfg = CN.load_yaml_with_base(URI("config://corner_detection/_base.yaml"))
```

```python
def show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
filename = "/Users/georg/chess_data/render/train/0064.png"
#filename = "/Users/georg/chess_data/render/train/2869.png"
filename = "/Users/georg/Desktop/b06.jpg"
filename = "/Users/georg/Desktop/IMG_0755.jpg"
img = cv2.imread(filename)
img, img_scale = resize_image(cfg, img)
show(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
```

```python
edges = detect_edges(cfg, gray)

plt.imshow(edges, cmap="gray")
```

```python
lines = detect_lines(cfg, edges)
all_horizontal_lines, all_vertical_lines = cluster_horizontal_and_vertical_lines(lines)

all_lines_img = img.copy()
draw_lines(all_lines_img, all_horizontal_lines, (0, 255, 0))
draw_lines(all_lines_img, all_vertical_lines, (255, 0, 0))
show(all_lines_img)
```

```python
horizontal_lines = eliminate_similar_lines(all_horizontal_lines, all_vertical_lines)
vertical_lines = eliminate_similar_lines(all_vertical_lines, all_horizontal_lines)

lines_img = img.copy()
draw_lines(lines_img, horizontal_lines, (0, 255, 0))
draw_lines(lines_img, vertical_lines, (255, 0, 0))
show(lines_img)
```

```python
intersection_points = get_intersection_points(horizontal_lines, vertical_lines)

show(lines_img)
plt.scatter(*intersection_points.T, c="r")
plt.title("intersection points")
```

```python
best_num_inliers = 0
best_configuration = None
iterations = 0
chosen_points = None
orig_scaled_warped_points = None
chosen_scaled_inlier_warped_points = None
while iterations < 30 or best_num_inliers < 30:
    row1, row2 = _choose_from_range(len(horizontal_lines))
    col1, col2 = _choose_from_range(len(vertical_lines))
    transformation_matrix = compute_homography(intersection_points,
                                                row1, row2, col1, col2)
    warped_points = warp_points(transformation_matrix, intersection_points)
    inlier_warped_points, inlier_intersection_points, horizontal_scale, vertical_scale = discard_outliers(cfg, warped_points, intersection_points)
    num_inliers = np.prod(inlier_warped_points.shape[:-1])
    if num_inliers > best_num_inliers:
        inlier_warped_points *= np.array((horizontal_scale, vertical_scale))

        # Quantize
        (xmin, xmax, ymin, ymax), scale, quantized_points, inlier_intersection_points, warped_img_size = quantize_points(cfg, inlier_warped_points, inlier_intersection_points)
        num_inliers = np.prod(quantized_points.shape[:-1])

        if num_inliers > best_num_inliers:
            best_num_inliers = num_inliers
            best_configuration = (xmin, xmax, ymin, ymax), scale, quantized_points, inlier_intersection_points, warped_img_size
            chosen_points = row1, row2, col1, col2
            orig_scaled_warped_points = warped_points * np.array((horizontal_scale, vertical_scale))
            chosen_scaled_inlier_warped_points = inlier_warped_points
    iterations += 1
    if iterations > 1000:
        print("too many iterations; stopped")
        break


# Retrieve best configuration
orig_intersection_points = intersection_points
(xmin, xmax, ymin, ymax), scale, quantized_points, intersection_points, warped_img_size = best_configuration
row1, row2, col1, col2 = chosen_points
```

```python
chosen_points = np.stack([
    orig_intersection_points[row1, col1],
    orig_intersection_points[row2, col1],
    orig_intersection_points[row2, col2],
    orig_intersection_points[row1, col2]
], axis=0)

show(lines_img)
plt.scatter(*orig_intersection_points.T, c="r", label="outliers")
plt.scatter(*intersection_points.T, c="g", label="inliers")
plt.scatter(*chosen_points.T, c="b", label="chosen sample")
plt.title("RANSAC result")
plt.legend()
```

```python
plt.scatter(*orig_scaled_warped_points.T, c="r", label="outliers")
plt.scatter(*chosen_scaled_inlier_warped_points.T, c="g", label="inliers")
plt.title("intersection points in scaled warped space")
plt.legend()
plt.axis("equal")
xticks = np.arange(np.floor(orig_scaled_warped_points[..., 0].min()), np.ceil(orig_scaled_warped_points[..., 0].max()))
yticks = np.arange(np.floor(orig_scaled_warped_points[..., 1].min()), np.ceil(orig_scaled_warped_points[..., 1].max()))
plt.xticks(xticks)
plt.yticks(yticks)
plt.grid()
```

```python
# Recompute transformation matrix based on all inliers
transformation_matrix = compute_transformation_matrix(
    intersection_points, quantized_points)
inverse_transformation_matrix = np.linalg.inv(transformation_matrix)

# Warp grayscale image
warped = cv2.warpPerspective(
    gray, transformation_matrix, tuple(warped_img_size.astype(np.int)))

plt.imshow(warped, cmap="gray")
```

```python
G_x = np.abs(cv2.Sobel(warped, cv2.CV_64F, 1, 0, ksize=cfg.BORDER_REFINEMENT.SOBEL_KERNEL_SIZE))
G_y = np.abs(cv2.Sobel(warped, cv2.CV_64F, 0, 1, ksize=cfg.BORDER_REFINEMENT.SOBEL_KERNEL_SIZE))

plt.figure()
plt.title("Sobel filter (horizontal)")
plt.imshow(G_x)
plt.colorbar()

plt.figure()
plt.title("Sobel filter (vertical)")
plt.imshow(G_y)
plt.colorbar()
```

```python
# Refine board boundaries
xmin, xmax = compute_vertical_borders(cfg, warped, scale, xmin, xmax)
ymin, ymax = compute_horizontal_borders(cfg, warped, scale, ymin, ymax)
```

```python
# Transform boundaries to image space
corners = np.array([[xmin, ymin],
                    [xmax, ymin],
                    [xmax, ymax],
                    [xmin, ymax]]).astype(np.float)
corners = corners * scale
img_corners = warp_points(inverse_transformation_matrix, corners)

show(img)
plt.scatter(*img_corners.T, c="r")
plt.title("predicted corners")
```
